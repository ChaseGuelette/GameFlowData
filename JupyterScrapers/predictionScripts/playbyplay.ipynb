```python
from dotenv import load_dotenv
import os
import pandas as pd
from nba_api.stats.endpoints import leaguedashplayerclutch
import time
from datetime import datetime, timedelta
from sqlalchemy import create_engine

# Load environment variables (database connection)
load_dotenv()
DATABASE_URL = os.getenv("DATABASE_URL")
engine = create_engine(DATABASE_URL)

def scrape_clutch(num_seasons=5):
    """
    Fetches clutch stats for the last `num_seasons` seasons and stores them in a database.
    
    Args:
        num_seasons (int): Number of past seasons to scrape. Default is 5 seasons.
    
    Returns:
        final_df (DataFrame): Combined dataframe of all clutch stats.
    """
    all_clutch_data = []  # List to store data
    
    current_year = datetime.now().year  # Get current year
    start_season = current_year - num_seasons  # Determine starting season

    for season_start in range(start_season, current_year):  # Loop through X seasons
        season = f"{season_start}-{str(season_start+1)[-2:]}"  # Format season (e.g., '2019-20')

        # Define the start and end date for each season
        start_date = datetime(season_start, 10, 24)  # Example: October 24th
        end_date = datetime(season_start + 1, 4, 14)  # Example: April 14th (end of season)

        print(f"\n🔍 Fetching clutch stats for season: {season}")

        # Loop through the season in 5-day increments
        current_date = start_date
        while current_date <= end_date:
            next_date = current_date + timedelta(days=4)  # 5-day range
            date_from = current_date.strftime('%m/%d/%Y')
            date_to = next_date.strftime('%m/%d/%Y')

            print(f"📅 Fetching clutch stats from {date_from} to {date_to} for {season}...")

            try:
                # Fetch clutch stats for the 5-day period
                clutch_stats = leaguedashplayerclutch.LeagueDashPlayerClutch(
                    season=season,
                    date_from_nullable=date_from,
                    date_to_nullable=date_to,
                    clutch_time='Last 5 Minutes',
                    ahead_behind='Ahead or Behind',
                    point_diff='5'
                )

                clutch_df = clutch_stats.get_data_frames()[0]  # Convert to DataFrame
                clutch_df['SEASON'] = season  # Add season column
                clutch_df['DATE_FROM'] = date_from  # Add date range
                clutch_df['DATE_TO'] = date_to

                # Compute additional clutch stats
                clutch_df['CLUTCH_SCORE_PCT'] = (clutch_df['FGM'] / clutch_df['FGA']).fillna(0)
                clutch_df['CLUTCH_USAGE_RATE'] = (clutch_df['FGA'] + 0.44 * clutch_df['FTA'] + clutch_df['TOV']) / clutch_df['MIN']
                clutch_df['CLUTCH_NET_RATING'] = clutch_df['PLUS_MINUS'] / clutch_df['GP']

                # Append data to list
                all_clutch_data.append(clutch_df)

                # Store data in the database (append instead of replace)
                with engine.connect() as conn:
                    with conn.begin():
                        clutch_df.to_sql('advanced_player_clutch_stats_by_date', conn, if_exists='append', index=False)

            except Exception as e:
                print(f"⚠️ Error fetching data for {season} ({date_from} - {date_to}): {e}")

            # Move to the next 5-day period
            current_date = next_date + timedelta(days=1)  # Start from next day

            # Sleep to avoid rate limiting
            time.sleep(2)

    # Combine all data into a single DataFrame
    final_df = pd.concat(all_clutch_data, ignore_index=True)

    # Save as CSV for backup
    final_df.to_csv('clutch_stats_multiple_seasons.csv', index=False)

    print("\n✅ Clutch stats collected for all seasons!")
    return final_df

# Run the function for the last 5 seasons
scrape_clutch(num_seasons=5)
```

```python
from nba_api.stats.endpoints import leaguedashplayerstats, playbyplayv2
import pandas as pd
import numpy as np
from sklearn import linear_model

def collect_advanced_player_data(cutoff_date):
    """Collect advanced player stats up to a specific cutoff date"""
    
    # Get advanced metrics
    adv_stats = leaguedashplayerstats.LeagueDashPlayerStats(
        measure_type_detailed_defense='Advanced',
        per_mode_detailed='PerGame',
        date_to_nullable=cutoff_date,  # Historical cutoff
        season='2023-24'
    )
    
    adv_df = adv_stats.get_data_frames()[0]
    
    # Get scoring metrics
    scoring_stats = leaguedashplayerstats.LeagueDashPlayerStats(
        measure_type_detailed_defense='Scoring',
        per_mode_detailed='PerGame',
        date_to_nullable=cutoff_date,
        season='2023-24'
    )
    
    scoring_df = scoring_stats.get_data_frames()[0]
    
    # Merge datasets
    # player_data = pd.merge(adv_df, scoring_df, on='PLAYER_ID')

    # Identify common columns between the two dataframes (metadata columns)
    common_cols = list(set(adv_df.columns).intersection(set(scoring_df.columns)))
    
    # Keep track of non-common columns from each dataframe
    adv_unique_cols = [col for col in adv_df.columns if col not in common_cols or col == 'PLAYER_ID']
    scoring_unique_cols = [col for col in scoring_df.columns if col not in common_cols or col == 'PLAYER_ID']
    
    # Merge on common columns to avoid duplicates
    # This ensures we only get one copy of each common column
    player_data = pd.merge(
        adv_df[adv_unique_cols],
        scoring_df[scoring_unique_cols],
        on='PLAYER_ID'
    )
    
    # Add back other common columns from either dataframe (they're the same)
    for col in common_cols:
        if col != 'PLAYER_ID':  # We already used PLAYER_ID as the merge key
            player_data[col] = adv_df[col]
    
    return player_data
```

```python
def estimate_offensive_rating(player_data, play_by_play_data):
    """
    Build an estimated offensive rating model
    
    This combines traditional offensive rating with regression
    adjustments based on lineup data
    """
    # Start with raw offensive rating
    player_data['E_OFF_RATING_BASE'] = player_data['OFF_RATING']
    
    # Add adjustments based on teammate quality
    # (This would use play-by-play data to assess lineups)
    
    # Create features for regression model
    features = ['USG_PCT', 'TS_PCT', 'AST_PCT', 'AST_TO', 'PACE', 'PIE']
    
    # Remove rows with missing values
    model_data = player_data.dropna(subset=features + ['OFF_RATING'])
    
    # Only include players with meaningful minutes
    model_data = model_data[model_data['MIN'] >= 10]
    
    # Build regression model
    X = model_data[features]
    y = model_data['OFF_RATING']
    
    # Train model
    reg = linear_model.Ridge(alpha=0.5)
    reg.fit(X, y)
    
    # Apply model to all players
    X_all = player_data[features].fillna(0)
    player_data['OFF_RATING_PREDICTED'] = reg.predict(X_all)
    
    # Blend raw and predicted values (weighted average)
    # More minutes = more weight on raw values
    weights = np.minimum(player_data['MIN'] / 20, 1)
    
    player_data['E_OFF_RATING'] = (
        weights * player_data['OFF_RATING'] + 
        (1 - weights) * player_data['OFF_RATING_PREDICTED']
    )
    
    return player_data
```

```python
def estimate_defensive_rating(player_data, play_by_play_data):
    """Build an estimated defensive rating model"""
    
    # Start with raw defensive rating
    player_data['E_DEF_RATING_BASE'] = player_data['DEF_RATING']
    
    # Create features for defensive regression model
    features = ['DREB_PCT', 'STL', 'BLK', 'PACE', 'PIE']
    
    # Remove rows with missing values
    model_data = player_data.dropna(subset=features + ['DEF_RATING'])
    
    # Only include players with meaningful minutes
    model_data = model_data[model_data['MIN'] >= 10]
    
    # Build regression model
    X = model_data[features]
    y = model_data['DEF_RATING']
    
    reg = linear_model.Ridge(alpha=0.5)
    reg.fit(X, y)
    
    # Apply model to all players
    X_all = player_data[features].fillna(0)
    player_data['DEF_RATING_PREDICTED'] = reg.predict(X_all)
    
    # Blend raw and predicted values
    weights = np.minimum(player_data['MIN'] / 20, 1)
    
    player_data['E_DEF_RATING'] = (
        weights * player_data['DEF_RATING'] + 
        (1 - weights) * player_data['DEF_RATING_PREDICTED']
    )
    
    return player_data
```

```python
def calculate_net_rating(player_data):
    """Calculate estimated net rating"""
    
    # Net rating is offensive rating minus defensive rating
    player_data['E_NET_RATING'] = player_data['E_OFF_RATING'] - player_data['E_DEF_RATING']
    
    return player_data
```
