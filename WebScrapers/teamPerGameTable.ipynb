```python
from bs4 import BeautifulSoup, Comment
import requests

import pandas as pd
```

```python
# with open("mvp/1991.html", encoding='utf-8') as file:
#     page = file.read()

with open("ATL_stats.html", encoding='utf-8') as file:
    page = file.read()
    
soup = BeautifulSoup(page, 'html')
print(soup.prettify())
```

```python
layer1 = soup.find('div', id='wrap')
layer2 = layer1.find('div', id='all_per_game_stats')
print(layer2)
```

```python
table_container = layer2.find('div', id='div_per_game_stats')
table = table_container.find('table')

headers = table.find('thead')
headers = headers.find_all('th')
print(len(headers))
#print(headers)
list1 = []

for item in headers:
    text = item.text.strip()
    if any(c.isalnum() for c in text):
        list1.append(text)
print(list1)
print(len(list1))
```

```python
#now that headers are extracted, we are going to convert this to a dataframe
df = pd.DataFrame(columns = list1)
pd.set_option('display.max.columns', 40)
df
```

```python
tableData = table.find('tbody')
rows = tableData.find_all('tr')
#print(len(rows))

#this clears the table inthe event of running a jupyter cell twice
df = pd.DataFrame(columns=df.columns)

for row in rows:
    col1 = row.find('th')
    item_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in item_data]
    individual_row_data.insert(0,col1)
    
    # print(individual_row_data)
    # print(len(individual_row_data))
    # print(individual_row_data[0])
    
    length=len(df)
    df.loc[length] = individual_row_data
df
```
